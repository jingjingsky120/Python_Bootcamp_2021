### Question 1
**Write a function called `get_manager` that takes as its one argument the Pandas DataFrame "Employees" and returns a DataFrame containing all employees (EmployeeID, first name, middle name, last name), and their manager's first and last name. The columns in the output DataFrame should be: EmployeeID, FirstName, MiddleName, LastName, ManagerFirstName, ManagerLastName.**




```python
def get_manager(df):
    
    # Step 1: create two DFs for mappings of employeeID -> Manager ID and EmployeeID -> Employee Name
    # note that managers are also employees and hence manager IDs are also employee IDs
    employee_manager_df = df[["EmployeeID","FirstName","MiddleName","LastName","ManagerID"]]
    employee_id_name_df = df[["EmployeeID","FirstName","LastName"]]
    
    # Step 2: merge the two but make sure which name is for whom
    out = pd.merge(
        employee_manager_df
        , employee_id_name_df
        , left_on = 'ManagerID'
        , right_on = 'EmployeeID'
        , how = 'inner'  # i'm excluding those without managers but you can argue about that
        , suffixes = ('_Employee', '_Manager')  # optional but more explicit and easier to check for error 
    )
    
    # Step 3: cleaning
    
    out = out.drop(['ManagerID', 'EmployeeID_Manager'], axis=1)
    out.columns = ["EmployeeID","FirstName","MiddleName","LastName","ManagerFirstName","ManagerLastName"]
    
    return out
```

```python
df1 = get_manager(Employees)
print("Shape of resulting table: ", df1.shape)
print("Columns: ", ', '.join(df1.columns))
df1.head()
```

### data visualization

```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix, parallel_coordinates
import seaborn as sns
import matplotlib.pylab as plt
```

```python
housing_df = pd.read_csv('Housing.csv')
housing_df.head(9)
```

```python
## Set the color of the points in the scatterplot and draw as open circles.
plt.scatter(housing_df.LSTAT, housing_df.MEDV, color='C2', facecolor='none')
plt.xlabel('LSTAT'); plt.ylabel('MEDV')

## barchart of CHAS vs. mean MEDV
# compute mean MEDV per CHAS = (0, 1)
dataForPlot = housing_df.groupby('CHAS').mean().MEDV
## using the variable ax for single a Axes
fig, ax = plt.subplots()
ax.bar(dataForPlot.index, dataForPlot, color=['C5', 'C1'])
ax.set_xticks([0, 1], False)
ax.set_xlabel('CHAS')
ax.set_ylabel('Avg. MEDV')
```

```python
## histogram of MEDV
ax = housing_df.MEDV.hist()
ax.set_xlabel('MEDV'); ax.set_ylabel('count')

## boxplot of MEDV for different values of CHAS
ax = housing_df.boxplot(column='MEDV', by='CHAS')
ax.set_ylabel('MEDV')
plt.suptitle('')  # Suppress the titles
plt.title('boxplot for MEDV')

# alternative plot with matplotlib
dataForPlot = [list(housing_df[housing_df.CHAS==0].MEDV),
      list(housing_df[housing_df.CHAS==1].MEDV)]
fig, ax = plt.subplots()
ax.boxplot(dataForPlot)
ax.set_xticks([1, 2], False)
ax.set_xticklabels([0, 1])
ax.set_xlabel('CHAS'); ax.set_ylabel('MEDV')
plt.show()
```

```python
## simple heatmap of correlations (without values)
corr = housing_df.corr()

# Include information about values (example demonstrate how to control the size of
# the plot
fig, ax = plt.subplots()
fig.set_size_inches(11, 7)
sns.heatmap(corr, annot=True, fmt=".1f", cmap="RdBu", center=0, ax=ax)
```
